```python
from openai import OpenAI
client = OpenAI()

res = client.responses.create(
    model="gpt-4.1-mini",
    input="hello"
)

print(res.output_text)
```

```python
def call_llm(prompt, user_input):
    res = client.responses.create(
        model="gpt-4.1-mini",
        input=prompt + user_input
    )
    return res.output_text
```

```python
import streamlit as st

user_input = st.text_input("입력:")

if user_input:
    response = call_llm("You are a game master.", user_input)
    st.write(response)
```
